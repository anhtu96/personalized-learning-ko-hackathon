```python
import pandas as pd
import time
import tqdm
import numpy as np
```

```python
start = time.time()
df = pd.read_csv('datasets/KT3_merged.csv')
print(time.time() - start)
```

```python
df['normalize_timestamp'] = (df['timestamp'] / 86400000).astype(int)
```

```python
df
```

```python
df['action_type'].value_counts()
```

Important notes:

Num enter = num submit + num quit

--> When you enter, you either submit or you quit, so we know the time spend each time by using time submit - time enter before that.

# Check time spend everyday by using enter, submit and quit

```python
df = df[df['action_type'].isin(['enter', 'submit', 'quit'])]
```

```python
df.index = range(len(df))
```

```python
df
```

```python
df['time_session'] = df['timestamp'].diff().copy(deep=True)
```

```python
df
```

```python
df.to_csv('KT3_intermediate.csv', index=False)
```

```python
df = pd.read_csv('KT3_intermediate.csv')
```

```python
df = df[df.index % 2 != 0]
```

```python
df
```

```python
df_groupby_result = df.groupby(['user_id', 'normalize_timestamp'],as_index=False)['time_session'].sum()
```

```python
df_groupby_result['time_spend'] = (df_groupby_result['time_session'] / 1000).astype('int64')
```

```python
df_groupby_result['day'] = df_groupby_result['normalize_timestamp']
```

```python
df_groupby_result
```

```python
df_groupby_result = df_groupby_result.drop(['normalize_timestamp', 'time_session'], axis=1)
```

```python
df_groupby_result = df_groupby_result[['user_id', 'day', 'time_spend']]
```

```python
df_groupby_result.to_csv('KT3_timespend_user_each_day.csv', index=False)
```
