# Load KT3 intermediate and join with question

```python
df = pd.read_csv('KT3_intermediate.csv')
```

```python
df
```

## Get the "time_spend_with_answer"

```python
list_of_quit_index = [x for x in df[df['action_type'] == 'quit'].index.values]
list_of_enter_quit_index = list([int(x - 1) for x in list_of_quit_index]) + list(list_of_quit_index)
```

```python
len(list_of_enter_quit_index)
```

```python
df_with_answer = df[~df.index.isin(list_of_enter_quit_index)].reset_index(drop=True)
```

```python
df_with_answer['time_session'] = df_with_answer['timestamp'].diff().copy(deep=True)
```

```python
df_with_answer
```

```python
df_with_answer = df_with_answer[df_with_answer.index % 2 != 0]
```

```python
df_with_answer
```

```python
df_with_answer['time_spend'] = (df_with_answer['time_session'] / 1000).astype('int64')
```

```python
df_with_answer
```

```python
df_with_answer.to_csv('KT3_intermediate_2.csv', index=False)
```
